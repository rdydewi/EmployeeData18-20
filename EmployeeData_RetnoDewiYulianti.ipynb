```python
#Imputing Categorical Variables
data_20['Gender'] = data_20['Gender'].fillna(data_20['Gender'].mode()[0])
data_20['Position'] = data_20['Position'].fillna(data_20['Position'].mode()[0])
data_20['Total years of experience'] = data_20['Total years of experience'].fillna(data_20['Total years of experience'].mode()[0])
data_20['Seniority level'] = data_20['Seniority level'].fillna(data_20['Seniority level'].mode()[0])
data_20['Years of experience in Germany'] = data_20['Years of experience in Germany'].fillna(data_20['Years of experience in Germany'].mode()[0])
data_20['Your main technology / programming language'] = data_20['Your main technology / programming language'].fillna(data_20['Your main technology / programming language'].mode()[0])
data_20['Other technologies/programming languages you use often'] = data_20['Other technologies/programming languages you use often'].fillna(data_20['Other technologies/programming languages you use often'].mode()[0])
data_20['Yearly bonus + stocks in EUR'] = data_20['Yearly bonus + stocks in EUR'].fillna(data_20['Yearly bonus + stocks in EUR'].mode()[0])
data_20['Annual bonus+stocks one year ago. Only answer if staying in same country'] = data_20['Annual bonus+stocks one year ago. Only answer if staying in same country'].fillna(data_20['Annual bonus+stocks one year ago. Only answer if staying in same country'].mode()[0])
data_20['Number of vacation days'] = data_20['Number of vacation days'].fillna(data_20['Number of vacation days'].mode()[0])
data_20['Employment status'] = data_20['Employment status'].fillna(data_20['Employment status'].mode()[0])
data_20['?ontract duration'] = data_20['?ontract duration'].fillna(data_20['?ontract duration'].mode()[0])
data_20['Main language at work'] = data_20['Main language at work'].fillna(data_20['Main language at work'].mode()[0])
data_20['Company size'] = data_20['Company size'].fillna(data_20['Company size'].mode()[0])
data_20['Company type'] = data_20['Company type'].fillna(data_20['Company type'].mode()[0])
data_20['Have you lost your job due to the coronavirus outbreak?'] = data_20['Have you lost your job due to the coronavirus outbreak?'].fillna(data_20['Have you lost your job due to the coronavirus outbreak?'].mode()[0])
data_20['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'] = data_20['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'].fillna(data_20['Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'].mode()[0])
```
